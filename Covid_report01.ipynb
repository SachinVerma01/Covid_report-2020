```python
Covid_data2=Covid_data1[Covid_data1["Confirmed"] != 0]
```

```python
Covid_data2
```

```python
x=Covid_data2.groupby(by="Country/Region")
df_1=pd.DataFrame(x.Confirmed.sum().sort_values()).reset_index()
df_1.sort_values(by="Confirmed",ascending=False,inplace=True)
df_1.columns=["Country","Cases"]
```

```python
df_1
```

# Name of country which are most affected by the covid 

```python
plt.figure(figsize=(22,12))
plt.bar(df_1["Country"][0:30],df_1["Cases"][0:30])
plt.xlabel("Country",fontdict={"c":"k","size":"20"})
plt.ylabel("Number of Covid Cases",fontdict={"c":"k","size":"20"})
plt.xticks(rotation="vertical")
plt.show()
```

```python
Covid_data1
```

```python
Covid_data3=Covid_data1[Covid_data1["Deaths"] != 0]
```

```python
x=Covid_data3.groupby(by="Country/Region")
df_2=pd.DataFrame(x.Deaths.sum()).reset_index()
df_2.columns=["Country","Death"]
df_2.sort_values(by="Death",ascending= False,inplace=True)
```

# Most number of recovered in which country

```python
Covid_data3=Covid_data1[Covid_data1["Recovered"] != 0]
x=Covid_data3.groupby(by="Country/Region")
df_3=pd.DataFrame(x.Recovered.sum()).reset_index()
df_3.columns=("Country","Recovered")
df_3.sort_values(by="Recovered",ascending=False,inplace=True)
```

```python
Covid_data2["Date"]=pd.to_datetime(Covid_data2["Date"])
```

```python
Covid_data2.info()
```

```python
Covid_data2["Month"]=Covid_data2["Date"].dt.month
Covid_data2["year"]=Covid_data2["Date"].dt.year
```

```python
Covid_data2
```

```python
Covid_data2["Month"].value_counts()
```

```python
Covid_data6=Covid_data2[Covid_data2["Confirmed"] != 0]
```

```python
x=Covid_data6.groupby(by="Month")
df_6=pd.DataFrame(x.Confirmed.sum()).reset_index()
df_6.columns=["month","confirmed"]
df_6.sort_values(by="month",ascending=True,inplace=True)
```

```python
Covid_data7=Covid_data2[Covid_data2["Deaths"] != 0]
```

```python
x=Covid_data6.groupby(by="Month")
df_7=pd.DataFrame(x.Deaths.sum()).reset_index()
df_7.columns=["month","Death"]
df_7.sort_values(by="month",ascending=True,inplace=True)
```

```python
Covid_data8=Covid_data2[Covid_data2["Recovered"] != 0]
```

```python
x=Covid_data8.groupby(by="Month")
df_8=pd.DataFrame(x.Recovered.sum()).reset_index()
df_8.columns=["month","Recovered"]
df_8.sort_values(by="month",ascending=True,inplace=True)
```

```python
plt.figure(figsize=(30,15))
plt.plot(df_6["month"],df_6["confirmed"],marker="p",markerfacecolor="r",markeredgecolor="k",markersize=11,color="darkblue")
plt.plot(df_7["month"],df_7["Death"],marker="p",markerfacecolor="r",markeredgecolor="k",markersize=11,color="red")
plt.plot(df_8["month"],df_8["Recovered"],marker="p",markerfacecolor="r",markeredgecolor="k",markersize=11,color="green")


for x,y in zip(df_8["month"],df_8["Recovered"]):
    plt.text(x,y,y,size=15,color="k")    

for x,y in zip(df_7["month"],df_7["Death"]):
    plt.text(x,y,y,size=15,color="k")

for x,y in zip(df_6["month"],df_6["confirmed"]):
    plt.text(x,y,y,size=15,color="k")
plt.legend(["Confirmed","Death","Recovered"])
plt.show()
```
